# 네이버 영화 인물 직업 테이블에 값 입력

```python
# MySQL Connector 모듈 및 pandas 모듈 import
import mysql.connector as mqc
import numpy as np
import pandas as pd
```

```python
# MySQL 연결 객체 생성
cnx2 = mqc.connect(host='v.pdmnu.com', user='root', password='!Bigdata10', database='naver-movie')

# MySQL을 실질적으로 실행하는 cursor 객체 생성
cursor2 = cnx2.cursor()
```

```python
# 실행할 쿼리문 입력
query = 'SELECT staff_part FROM `movie-staff`;'

# cursor 객체를 통하여 쿼리 실행
cursor2.execute(query)
```

```python
# 결과물을 저장할 리스트 선언
part_name = []

# 결과물을 리스트에 저장
# 한 개의 컬럼만 조회하였지만 결과물이 튜플들의 리스트로 반환되기 때문에 0번째 인덱스의 값만 가져오도록 코드 작성
for _ in cursor2:
    part_name.append(_[0])
part_name
```

```python
# 결과물을 저장할 Series 객체 생성
part_series = pd.Series(part_name, name='Part')
part_series
```

```python
# unique() 메서드를 활용하여 중복값을 제거한 후 반환된 ndarray 객체를 다시 Series 객체로 생성
# 이후 값을 기준으로 재정렬
part_series = pd.Series(part_series.unique(), name='Part')
part_series.sort_values(inplace=True, ignore_index=True)
part_series
```

```python
# 정리된 Series 객체의 값만 추출하여 MySQL 쿼리에 대입할 수 있도록 튜플들의 리스트로 변환
value_params = list(zip(part_series.T.values, ))
value_params
```

```python
# MySQL 쿼리 실행
query = 'INSERT INTO `professions` (prof_name) VALUES (%s)'
try:
    cursor2.executemany(query, value_params)
except mqc.Error as err:
    print('오류가 발생하였습니다:', err)
finally:
    print(str(cursor2.rowcount) + '개의 행이 처리되었습니다.')
    cnx2.commit()
```

```python
# MySQL 연결 닫기
cnx2.close()
```
